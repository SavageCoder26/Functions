6. Explain the concept of generators in Python and how they are defined.

Ans.A generator produces items one at a time and only when requested.Generators are iterators, so they implement __iter__() and __next__() internally.

Instead of return, use yield to produce a value without exiting the function. The function then becomes a generator.

  for example


      def simple_generator():
          yield 1
          yield 2
          yield 3

      gen = simple_generator()  # gen is a generator object

      print(next(gen))  # 1
      print(next(gen))  # 2
      print(next(gen))  # 3
      # next(gen) now raises StopIteration

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

ANs. 1. map()

        Definition:
        Applies a given function to each element of an iterable and returns a map object (iterator) with the results.

        Points:

        Used for element-wise transformation.

        Can work with one or more iterables.

        Example:

        nums = [1, 2, 3, 4]
        squared = list(map(lambda x: x**2, nums))
        print(squared)  # [1, 4, 9, 16]

        2. filter()

        Definition:
        Filters elements of an iterable using a function that returns True or False for each element.

        Points:

        Used for selecting elements based on a condition.

        Function must return a boolean.

        Example:

        nums = [1, 2, 3, 4, 5, 6]
        evens = list(filter(lambda x: x % 2 == 0, nums))
        print(evens)  # [2, 4, 6]

        3. reduce()

        Definition:
        Applies a function cumulatively to elements of an iterable, reducing it to a single value.

        Points:

        Belongs to functools module, must be imported.

        Useful for aggregation (sum, product, max, etc.).

        Example:

        from functools import reduce

        nums = [1, 2, 3, 4]
        product = reduce(lambda a, b: a * b, nums)
        print(product)  # 24

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_evens(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:   # check if even
            even_sum += num
    return even_sum


# Example usage
nums = [1, 2, 3, 4, 5, 6, 10, 15]
print(sum_of_evens(nums))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:   # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):  # check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True

# Checking from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is Prime")
    else:
        print(f"{num} is Not Prime")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms   # how many terms to generate
        self.count = 0               # current position
        self.a, self.b = 0, 1        # first two Fibonacci numbers

    def __iter__(self):
        return self   # iterator object returns itself

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration   # end iteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b  # update Fibonacci numbers
            self.count += 1
            return self.b

fib = FibonacciIterator(10)   # first 10 terms

for num in fib:
    print(num, end=" ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2 ** i

for value in powers_of_two(5):   # up to 2^5
    print(value, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    """
    Generator function that yields each line of a file one by one.
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # remove trailing newline

file_path = "example.txt"  # path to your file

for line in read_file_line_by_line(file_path):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Conversion function: C to F
def c_to_f(c):
    return (c * 9/5) + 32

# Using map to convert
fahrenheit = list(map(c_to_f, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Input string
text = "Hello Python World"

# Function to check if a character is NOT a vowel
def is_not_vowel(char):
    return char.lower() not in "aeiou"

# Using filter to remove vowels
result = ''.join(filter(is_not_vowel, text))

print(result)
```
